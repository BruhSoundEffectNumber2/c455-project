```python
def schwarzian(l: np.float64, x: np.float64) -> np.float64:
    """Schwarzian derivative of Hassell's map function."""
    return hprime3(l, x) / hprime(l, x) - 1.5 * (hprime2(l, x) / hprime(l, x)) ** 2
```

```python
print(schwarzian(2, 10))
```

```python
for l in [0.5, 1.0, 1.5, 10, 22, 30]:
    xs = np.concatenate([np.linspace(0, 0.09, 100), np.linspace(0.13, 10, 100)])
    ys = schwarzian(l, xs)
    plt.ylim(-300, 20)
    plt.title(f"lambda = {l}")
    sns.lineplot(x=xs, y=ys)
    plt.show()
    print(f"# ys > 0: for l={l}", np.sum(ys > 0))
```
