```python
def histogram(l, bins=100, binrange=None, n=10000):
    res = iterate_postrans(l, 1/9, n)[1]
    plt.figure(dpi=600)
    sns.histplot(res, bins=bins, binrange=binrange)
    plt.title(f"Histogram of Hassell where $\lambda$={l}")
    plt.xlabel("Value of x")
    plt.ylabel("Frequency")
```
